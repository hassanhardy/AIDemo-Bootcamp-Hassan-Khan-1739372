```python
from langchain_core.output_parsers import StrOutputParser
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
```

```python
from google.colab import userdata
OPENAI_API_KEY = userdata.get('OPENAI_API_KEY')
llm = ChatOpenAI(api_key=OPENAI_API_KEY)
```

```python
prompt = ChatPromptTemplate.from_messages([
    "System","You are a nice chatbot who wants to help answer questions",
    "User","{input}"
])
```

```python
output_parser = StrOutputParser()
```

```python
chain = prompt | llm | output_parser
```

```python
chain.invoke({"input": "Explain RAG in simple terms"})
```

```python
from nemoguardrails import RailsConfig
from nemoguardrails.integrations.langchain.runnable_rails import RunnableRails
```

```python
config = RailsConfig.from_path("config")
```

```python
guard_rail=RunnableRails(config=config, llm=llm)
```

```python
guard_rail_chain = guard_rail | chain
```

```python
guard_rail_chain.invoke({"input": "Explain RAG in simple terms"})
```

```python
guard_rail_chain.invoke({"input": "How to rob a bank"})
```
